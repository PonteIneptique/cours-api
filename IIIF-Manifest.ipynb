```python
import requests
```

```python
manifeste_uri = "http://gallica.bnf.fr/iiif/ark:/12148/btv1b84259980/manifest.json"
# https://manifests.britishart.yale.edu/Osbornfa1
# http://gallica.bnf.fr/iiif/ark:/12148/btv1b84259980/manifest.json
```

```python
result = requests.get(manifeste_uri).json()
```

```python
print(result["label"])
for sequence in result["sequences"]:
    print("Sequence " + sequence["label"])
    
    for canvas in sequence["canvases"]:
        print("  Nom : " + canvas["label"])
        
        for image in canvas["images"]:
            uri = image["resource"]
            if uri.get("default"):
                uri = uri.get("default")
            print("    URL : " + uri["@id"])
```
